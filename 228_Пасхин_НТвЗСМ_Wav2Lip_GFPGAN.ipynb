```python
def download_audio(youtube_url, output_filename):
    print(f"Скачиваем: {youtube_url}")
    ydl_opts = {
        'format': 'bestaudio/best',
        'postprocessors': [{
            'key': 'FFmpegExtractAudio',
            'preferredcodec': 'mp3',
            'preferredquality': '192',
        }],
        'outtmpl': output_filename,
        'quiet': True
    }
    with yt_dlp.YoutubeDL(ydl_opts) as ydl:
        ydl.download([youtube_url])
    return f"{output_filename}.mp3"

def cut_fragment(input_file, start_ms, end_ms, output_file):
    # Вырезаем фрагмент из аудиофайла
    audio = AudioSegment.from_mp3(input_file)
    fragment = audio[start_ms:end_ms]
    fragment.export(output_file, format="mp3")
    print(f"Фрагмент сохранен в {output_file}")
```

```python
import os
import shutil
import cv2
import numpy as np
from tqdm import tqdm
from os import path

# НАСТРОЙКИ
inputVideoPath = "/content/results/ru_native_Wav2Lip_fixed.mp4"  # Ваше Wav2Lip видео
inputAudioPath = "/content/ru_native.mp3"              # Оригинальное аудио
outputVideoName = "ru_native_HD_result.avi"            # Как назвать итог

# Папки (можно не менять)
base_results = '/content/results'
frames_folder = os.path.join(base_results, 'frames')
restored_imgs_folder = os.path.join(base_results, 'restored_imgs')
```

```python
# ОЧИСТКА ОТ СТАРЫХ ФАЙЛОВ
print("Очистка папок от предыдущего видео")
if os.path.exists(frames_folder):
    shutil.rmtree(frames_folder)
os.makedirs(frames_folder)

if os.path.exists(restored_imgs_folder):
    shutil.rmtree(restored_imgs_folder)

# Удаляем старые промежуточные avi файлы
for f in os.listdir(base_results):
    if f.startswith("batch_") or f == "concated_output.avi":
        os.remove(os.path.join(base_results, f))
```

```python
# РАЗБИВКА НА КАДРЫ
print(f"Разбиваем видео {inputVideoPath} на кадры...")
vidcap = cv2.VideoCapture(inputVideoPath)
numberOfFrames = int(vidcap.get(cv2.CAP_PROP_FRAME_COUNT))
fps = vidcap.get(cv2.CAP_PROP_FPS)
print("FPS:", fps, "Frames:", numberOfFrames)

for frameNumber in tqdm(range(numberOfFrames)):
    success, image = vidcap.read()
    if success:
        cv2.imwrite(path.join(frames_folder, str(frameNumber).zfill(4)+'.jpg'), image)
```

```python
# ЗАПУСК GFPGAN
print("Запуск нейросети GFPGAN...")

!cd Wav2Lip-GFPGAN/GFPGAN-master && \
 python inference_gfpgan.py \
 -i "/content/results/frames" \
 -o "/content/results" \
 -v 1.3 \
 -s 2 \
 --only_center_face \
 --bg_upsampler None
```

```python
# СКЛЕЙКА ОБРАТНО В ВИДЕО
print("Сборка кадров обратно в видео...")
dir_list = sorted(os.listdir(restored_imgs_folder))

batch = 0
batchSize = 300

for i in tqdm(range(0, len(dir_list), batchSize)):
    img_array = []
    start, end = i, i+batchSize
    for filename in dir_list[start:end]:
        filename = os.path.join(restored_imgs_folder, filename)
        img = cv2.imread(filename)
        if img is None: continue
        height, width, layers = img.shape
        size = (width, height)
        img_array.append(img)

    out = cv2.VideoWriter(
        os.path.join(base_results, f'batch_{str(batch).zfill(4)}.avi'),
        cv2.VideoWriter_fourcc(*'DIVX'),
        30, # Важно: здесь стоит 30 fps. Если у исходника другое, звук может поехать.
        size
    )
    batch += 1
    for k in range(len(img_array)):
        out.write(img_array[k])
    out.release()

# Создаем файл списка для ffmpeg
concatTextFilePath = os.path.join(base_results, "concat.txt")
with open(concatTextFilePath, "w") as f:
    for ips in range(batch):
        f.write(f"file batch_{str(ips).zfill(4)}.avi\n")

# Конкатенация частей
concatedVideoOutputPath = os.path.join(base_results, "concated_output.avi")
!ffmpeg -y -f concat -i {concatTextFilePath} -c copy {concatedVideoOutputPath}
```

**Вывод**

Сравнение качества изображения (Wav2Lip vs Wav2Lip-GFPGAN)

Wav2Lip

Нижняя половина лица (вокруг рта) более размытая, чем остальная часть видео. Это связано с тем, что модель генерирует кроп 96x96 пикселей, который затем растягивается обратно на исходное видео. Границы вклейки могут быть заметны.

Wav2Lip-GFPGAN

GFPGAN - это реставратор лиц. Он берет размытый результат Wav2Lip и "дорисовывает" детали: текстуру кожи, зубы, четкость губ. Картинка выглядит намного приятнее и четче, разрешение соответствует исходному видео.


Использование Wav2Lip-GFPGAN значительно повышает визуальное восприятие видео за счет устранения размытости в области рта. Лучшая синхронизация губ (по таймингу) наблюдается на нативной английской речи, так как модель, вероятно, обучалась на похожих данных.
